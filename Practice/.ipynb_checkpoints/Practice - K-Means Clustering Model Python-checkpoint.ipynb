```python
# Import the libraries to be able to import data --> do calculations --> visualize the data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset =  pd.read_csv('/Users/joshuamarquez/anaconda_projects/Practice/digital_behavior.csv')
dataset
```

```python
len(dataset)
```

```python
# .iloc[] slices the data when passing in an index
# Let's do sleep_hours (index 2) and social_media_time_min (index 4)
x = dataset.iloc[:,2:4].values
x
```

```python
# SciKit learn is very interesting, powerful, and should check it out!
from sklearn.cluster import KMeans
wcss = []
for i in range (1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++')
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

# Output
wcss
```

```python
plt.plot(range(1, 11), wcss)
```

```python
# Here, from the graph above, the ideal number of clusters is when the y-value begins to level off
# So up until 4 clusters, the graph seems to be decreasing a lot
# But after 4, the data seems to be decreasing less and less
# Therefore, the ideal number of clusters is 4!
```

```python
kmeans = KMeans(n_clusters = 4, init = 'k-means++')
y_kmeans = kmeans.fit_predict(x)
y_kmeans
# This will output the cluster number each data point was assigned to, so COOOL!
```

```python
# 0 (Cluster 1): 218 times
# 1 (Cluster 2): 206 times
# 2 (Cluster 3): 101 times
# 3 (Cluster 4): 75 times
```

```python
pd.concat([dataset, pd.DataFrame(y_kmeans)], axis = 1)
```

```python
# Making a full dot plot, Color coding the clusters

# [y_kmeans == a, b]
# a = column index of data we are filtering?
# b = column index of data we are plotting?
# s = size
# c = color
# alpha = visibility

# We have 4 clusters, so we call the function ".scatter()" 4 times
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 100, c = 'red', alpha = 0.3)
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 100, c = 'orange', alpha = 0.3)
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 100, c = 'teal', alpha = 0.3)
plt.scatter(x[y_kmeans == 3, 0], x[y_kmeans == 3, 1], s = 100, c = 'blue', alpha = 0.3)
```

```python
# This is really cool! I especially like K-Means clustering and understand it better now
# I can just think of the applications, maybe we can use it on focus and mood? Let's try that actually.
dataset
```

```python
# Step 1: Check the data and find two good column indexes you want to compare
    # Let's choose focus_score (index 5) and anxiety_level (index 7)
    # and let's slice the data up
x2 = dataset.iloc[:,5:7].values
x2
```

```python
# Step 2: Pass in the sliced data and output a line graph to determine the ideal number of clusters
    # Declare Within Cluster Sum of Squares array = wcss2[]
    # Here, our data is x2, so we'll pass that in
wcss2 = []
for i in range (1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++')
    kmeans.fit(x2)
    wcss2.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss2)
```

```python
# Step 3: Now that we've found our ideal number of clusters (4)
# We'll store it in a y_means variable
    # kmeans2
    # y_means2
    # x2

kmeans2 = KMeans(n_clusters = 4, init = 'k-means++')
y_kmeans2 = kmeans2.fit_predict(x2)
y_kmeans2
```

```python
# Step 4: Concatenate the k-means assigned cluster values to the end of our original data set
pd.concat([dataset, pd.DataFrame(y_kmeans2)], axis = 1)
```

```python
# Step 5: Making a full dot plot! And color code the clusters!

# [y_kmeans == a, b]
# a = column index of data we are filtering?
# b = column index of data we are plotting?
# s = size
# c = color
# alpha = visibility

# We have 4 clusters, so we call the function ".scatter()" 4 times
plt.scatter(x2[y_kmeans2 == 0, 0], x2[y_kmeans2 == 0, 1], s = 100, c = 'red', alpha = 0.3)
plt.scatter(x2[y_kmeans2 == 1, 0], x2[y_kmeans2 == 1, 1], s = 100, c = 'orange', alpha = 0.3)
plt.scatter(x2[y_kmeans2 == 2, 0], x2[y_kmeans2 == 2, 1], s = 100, c = 'teal', alpha = 0.3)
plt.scatter(x2[y_kmeans2 == 3, 0], x2[y_kmeans2 == 3, 1], s = 100, c = 'blue', alpha = 0.3)
plt.xlabel('Focus Score (larger value = more focused)')
plt.ylabel('Anxiety Level (larger value = more anxious)')
```
